# Fake thesis title creator

This script generates random thesis titles from thematical databanks and predefined syntaxical structures.

To be improved eventually, via refinement and addition of thematical databanks, refinement of generation via if/then/else conditions, and output of the results into files insteand of 'print'.

```python
# Déclaration des banques de données thématiques pour l'instant vides et import des librairies.
# Declaring empty thematical data banks and importing libraries.

import csv
import random

epoque = []
oeuvres = []
citations = []
genres_txt = []
emotions = []
metiers = []
villes = []
prenoms = []
noms = []

# Récupération des expressions de dates et titres d'œuvres depuis le tableau originel enrichi des expressions extraites.
# Recording date expressions and work titles from the original table, enriched with extracted expressions.

with open('path/to/file') as csvalire:
    csvlu = csv.DictReader(csvalire, delimiter='\t')
    
    # Nécessite une colonne "MC_oeuvre" contenant des titres d'œuvres et une colonne "MC_date" contenant des expressions d'époques en str.
    # Needs an "MC_oeuvre" column with work titles and an "MC_date" column with expressions of dates as str.
    
    for line in csvlu:
        if line['MC_oeuvre'] != "":
            oeuvres.append(line['MC_oeuvre'])
        if line['MC_date'] != "":
            epoque.append(line['MC_date'])
        prenoms.append(line['author_firstname'])
        noms.append(line['author_name'])
```

```python
# Récupération des émotions depuis le tableau produit par WikiData (requête SPARQL), avec une colonne "itemLabel" contenant des noms d'émotions.
# Getting the emotions from the WikiData table (SPARQL request), with an "itemLabel" column containing names of emotions.

with open('path/to/file') as emotionsalire:
    emotionslues = csv.DictReader(emotionsalire)
    
    for line in emotionslues:
        emotions.append(line['itemLabel'])
```

```python
# Récupération des citations depuis le tableau produit par WikiData (requête SPARQL), avec une colonne "itemLabel" contenant des citations.
# Getting the quotes from the WikiData table (SPARQL request), with an "itemLabel" column containing quotes.

with open('path/to/file') as citalire:
    citlues = csv.DictReader(citalire)
    
    for line in citlues:
        if line['itemLabel'][0] != "Q":
            citations.append(line['itemLabel'])
```

```python
# Récupération des noms de métiers depuis le tableau produit par WikiData (requête SPARQL), avec une colonne "itemLabel" contenant des noms de métiers.
# Getting the job names from the WikiData table (SPARQL request), with an "itemLabel" column containing names of jobs.

with open('path/to/file') as metieralire:
    metierlues = csv.DictReader(metieralire)
    
    for line in metierlues:
        if line['itemLabel'][0] != "Q":
            metiers.append(line['itemLabel'])
```

```python
# Récupération des émotions depuis le tableau produit par WikiData (requête SPARQL), avec une colonne "itemLabel" contenant des noms d'émotions.
# Getting the emotions from the WikiData table (SPARQL request), with an "itemLabel" column containing names of emotions.

with open('path/to/file') as vilalire:
    villues = csv.DictReader(vilalire)
    
    for line in villues:
        if line['itemLabel'][0] != "Q":
            villes.append(line['itemLabel'])
```

```python
# Récupération des villes depuis le tableau produit par WikiData (requête SPARQL), avec une colonne "itemLabel" contenant des noms de villes.
# Getting the cities from the WikiData table (SPARQL request), with an "itemLabel" column containing names of cities.

with open('path/to/file') as genralire:
    genrlues = csv.DictReader(genralire)
    
    for line in genrlues:
        if line['itemLabel'][0] != "Q":
            genres_txt.append(line['itemLabel'])
```

```python
def make_title_1(epoque, citations, genres_txt, emotions, villes):
    """
    Fonction générant un titre selon une première structure,
    prenant en arguments les banques de données nécessaires.
    --------
    Function generating a title according to a first structure,
    taking as arguments the databanks it needs.
    """
    citation = random.choice(citations)
    sujet1 = random.choice(genres_txt)
    sujet2 = random.choice(emotions)
    localite = random.choice(villes)
    date = random.choice(epoque)
    
    print('"' + citation + '" : ' + sujet1 + " et " + sujet2 + " dans la " + localite + " " + date + ".")
    
def make_title_2(epoque, metiers, villes):
    """
    Fonction générant un titre selon une deuxième structure,
    prenant en arguments les banques de données nécessaires.
    --------
    Function generating a title according to a second structure,
    taking as arguments the databanks it needs.
    """
    sujet1 = random.choice(metiers)
    localite = random.choice(villes)
    date = random.choice(epoque)
    
    print('Être ' + sujet1 + " à "+ localite.replace("ville de ","") + " " + date + ".")

def make_title_3(epoque, citations, emotions):
    """
    Fonction générant un titre selon une troisième structure,
    prenant en arguments les banques de données nécessaires.
    --------
    Function generating a title according to a third structure,
    taking as arguments the databanks it needs.
    """
    citation = random.choice(citations)
    sujet1 = random.choice(emotions)
    date = random.choice(epoque)
    
    print('"' + citation + '" : ' + sujet1 + " " + date + ".")


# Génération des titres.
# Title generation.
counter = 1

while counter <= 40:
    counter += 1
    make_title_1(epoque, citations, genres_txt, emotions, villes)
    make_title_2(epoque, metiers, villes)
    make_title_3(epoque, citations, emotions)
```

```python
def make_names(noms, prenoms):
    """
    Fonction générant un faux nom de chartiste,
    prenant en arguments les banques de données nécessaires.
    --------
    Function generating a fake author name,
    taking as arguments the databanks it needs.
    """
    prenom = random.choice(prenoms)
    nom = random.choice(noms)
    
    print(nom + ", " + prenom)

# Génération des noms.
# Name generation.
counter = 0
while counter <= 250:
    counter += 1
    make_names(noms, prenoms)
```
